### Register Tables as SQL tables to perform SQL operations on our data

#### Create Database for each Layer

```python
spark.sql("""
USE CATALOG hive_metastore
""")

# Bronze database
spark.sql("""
CREATE DATABASE IF NOT EXISTS bronze
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/bronze/bronze_db'
""")

# Silver database
spark.sql("""
CREATE DATABASE IF NOT EXISTS silver
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/silver/silver_db'
""")

# Gold database
spark.sql("""
CREATE DATABASE IF NOT EXISTS gold
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/gold/gold_db'
""")
```

#### Register Bronze Tables in Bronze Database

```python
#Bronze tables
tables_bronze = [
    "orders",
    "customers",
    "geolocation",
    "order_items",
    "order_payments",
    "order_reviews",
    "products",
    "sellers"
]

for table in tables_bronze:
    spark.sql(f"""
              create TABLE if not exists bronze.{table}
              using delta 
              location 'abfss://olist-data@retailds.dfs.core.windows.net/bronze/{table}' """)
```

```python
# Confirm that tables are registered in database 
spark.sql("""show tables in bronze""").display()
```

#### Register Silver Tables in Silver Database

```python
# Silver Tables
tables_silver = [
     "orders",
    "customers",
    "order_items",
    "payments",
    "order_reviews",
    "products",
    "sellers"
]

for table in tables_silver:
    spark.sql(f"""
              create TABLE if not exists silver.{table}
              using delta 
              location 'abfss://olist-data@retailds.dfs.core.windows.net/silver/{table}' """)
```

```python
# Confirm that tables are registered in database 
spark.sql("""show tables in silver""").display()
```

#### Register Silver Tables in Gold Database

```python
# Gold Tables
tables_gold=[
    "fact_orders",
    "dim_date",
    "dim_customers",
    "dim_products",
    "dim_sellers",
    "fact_order_items",
    "fact_payments",
    "fact_order_reviews"
]

for table in tables_gold:
    spark.sql(f"""
              create table if not exists gold.{table}
              using delta
              location 'abfss://olist-data@retailds.dfs.core.windows.net/gold/{table}' """)
```

```python
# Confirm that tables are registered in database 
spark.sql("""show tables in gold""").display()
```
