### Register Tables as SQL tables to perform SQL operations on our data

#### Create Database for each Layer

```python
spark.sql("""
USE CATALOG hive_metastore
""")

# Bronze database
spark.sql("""
CREATE DATABASE IF NOT EXISTS bronze
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/bronze/bronze_db'
""")

# Silver database
spark.sql("""
CREATE DATABASE IF NOT EXISTS silver
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/silver/silver_db'
""")

# Gold database
spark.sql("""
CREATE DATABASE IF NOT EXISTS gold
LOCATION 'abfss://olist-data@retailds.dfs.core.windows.net/gold/gold_db'
""")
```

#### Register Bronze Tables in Branze Database

```python
#Bronze tables
tables = [
    "orders",
    "customers",
    "geolocation",
    "order_items",
    "order_payments",
    "order_reviews",
    "products",
    "sellers"
]

for table in tables:
    spark.sql(f"""
              create TABLE if not exists bronze.{table}
              using delta 
              location 'abfss://olist-data@retailds.dfs.core.windows.net/bronze/{table}' """)
```

```python
# Confirm that tables are registered in database 
spark.sql("""show tables in bronze""").display()
```
