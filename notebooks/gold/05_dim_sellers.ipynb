```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.sellers")
```

```python
dim_sellers = (
    spark.table("silver.sellers")
    .select(
        "seller_id",
        "seller_city",
        "seller_state"
    )
)
```

```python
dim_sellers.select(
    count("*").alias("total_rows"),
    countDistinct("seller_id").alias("distinct_sellers")
).show()
```

```python
dim_sellers.write\
    .format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/dim_sellers")
```
