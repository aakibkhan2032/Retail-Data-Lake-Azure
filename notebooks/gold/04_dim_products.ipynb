```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.products")
```

```python
dim_products = (
    spark.table("silver.products")
    .select(
        "product_id",
        "product_category_name",
        "product_name_lenght",
        "product_description_lenght",
        "product_photos_qty"
    )
)
```

```python
dim_products.select(
    count("*").alias("total_rows"),
    countDistinct("product_id").alias("distinct_products")
).show()
```

```python
dim_products.write\
    .format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/dim_products")
```
