```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.payments")
```

```python
fact_payments = (
    spark.table("silver.payments")
    .select(
        "order_id",
        "payment_sequential",
        "payment_type",
        "payment_installments",
        "payment_value"
    )
)
```

```python
fact_payments.select(
    count("*").alias("total_rows"),
    countDistinct("order_id", "payment_sequential").alias("distinct_payments")
).show()
```

```python
fact_payments.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/fact_payments")
```
