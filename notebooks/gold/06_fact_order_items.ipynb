```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.order_items").display()
```

```python
fact_order_items = (
    spark.table("silver.order_items")
    .select(
        "order_id",
        "order_item_id",
        "product_id",
        "seller_id",
        "shipping_limit_date",
        "price",
        "freight_value"
    )
    .withColumn(
        "total_item_amount",
        col("price") + col("freight_value")
    )
)
```

```python
fact_order_items.select(
    count("*").alias("total_rows"),
    countDistinct("order_id", "order_item_id").alias("distinct_items")
).show()
```

```python
fact_order_items.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/fact_order_items")
```
