```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.customers").display()
```

```python
dim_customers = (
    spark.table("silver.customers").select(
        "customer_id",
        "customer_unique_id",
        "customer_city",
        "customer_state"
    )

)
```

```python
dim_customers.select(
    countDistinct("customer_id").alias("distinct_customers"),
    count("*").alias("total_rows")
).show()
```

```python
dim_customers.write\
    .format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/dim_customers")
```
