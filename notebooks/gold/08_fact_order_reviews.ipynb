```python
from pyspark.sql.functions import *
```

```python
spark.table("silver.order_reviews")
```

```python
fact_order_reviews = (
    spark.table("silver.order_reviews")
    .select(
        "review_id",
        "order_id",
        "review_score",
        "review_comment_title",
        "review_comment_message",
        "review_creation_date"
    )
)
```

```python
fact_order_reviews.select(
    count("*").alias("total_rows"),
    countDistinct("order_id", "review_id").alias("distinct_reviews")
).show()
```

```python
fact_order_reviews.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .save("abfss://olist-data@retailds.dfs.core.windows.net/gold/fact_order_reviews")
```
