```python
# import necessary libraries
from pyspark.sql.functions import *
```

```python
df_order_payments = spark.read.format("delta").load("abfss://olist-data@retailds.dfs.core.windows.net/bronze/order_payments")
```

```python
df_order_payments.printSchema()
```

```python
df_order_payments.limit(5).display()
```

```python
df_order_payments.count()
```

```python
df_order_payments.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df_order_payments.columns                      
]).display()
```

```python
df_order_payments = df_order_payments.withColumn("payment_value", col("payment_value").cast("double"))\
    .withColumn("payment_installments", col("payment_installments").cast("integer"))
```

```python
df_order_payments.groupBy("order_id","payment_type","payment_installments","payment_value")\
    .count().display()
```

```python
df_order_payments.groupBy("order_id").count().filter(col("count").cast("integer")>1).orderBy(desc("count")).display()
```
