```python
def export_gold_table(df, table_name):
    (
        df
        .coalesce(1)  # single CSV file
        .write
        .mode("overwrite")
        .option("header", "true")
        .csv(f"{export_base_path}/{table_name}")
    )
```

```python
export_gold_table(df_dim_date, "dim_date")
export_gold_table(df_dim_customers, "dim_customers")
export_gold_table(df_dim_products, "dim_products")
export_gold_table(df_dim_sellers, "dim_sellers")
```

```python
export_gold_table(df_fact_orders, "fact_orders")
export_gold_table(df_fact_order_items, "fact_order_items")
export_gold_table(df_fact_payments, "fact_payments")
export_gold_table(df_fact_order_reviews, "fact_order_reviews")
```
